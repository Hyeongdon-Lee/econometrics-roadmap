## **파트 1: 회귀분석의 기초와 원리 (OLS Regression)**

- **1.** OLS (Ordinary Least Squares, 최소자승법)
- **2.** 누락 변수 편향 (Omitted Variable Bias, OVB)
- **3.** 가설 검정 (Hypothesis Testing)
- **4.** 강건 표준오차 (Robust Standard Errors)
- **5.** 다중공선성 (Multicollinearity)
- **6.** OLS 응용 기술: 더미 변수, 비선형 관계, 상호작용 항

## **파트 2: 전통적 인과추론 방법론 (Traditional Causal Inference)**

- **7.** 인과추론의 기본 원칙
- **8.** 무작위 통제 실험 (Randomized Controlled Trial, RCT)
- **9.** 이중차분법 (Difference-in-Differences, DID)
- **10.** 시차를 둔 이중차분법 (Staggered DID)
- **11.** 회귀 불연속 설계 (Regression Discontinuity Design, RDD)
- **12.** 도구변수법 (Instrumental Variables, IV)
- **13.** 고정효과 모형 (Fixed Effects Model)
- **14.** 합성 통제법 (Synthetic Control Method, SCM)
- **15.** 이중 강건 추정법 (Doubly Robust Estimation, DRE)

## **파트 3: 전통적 예측 & 시계열 방법론 (Traditional Prediction & Time Series)**

- **16.** 예측 & 시계열 분석의 기본 원칙
- **17.** AR (자기회귀) 모형
- **18.** MA (이동평균) 모형
- **19.** ARIMA 모형
- **20.** VAR (벡터 자기회귀) 모형
- **21.** 단위근과 공적분 (Unit Root and Cointegration)
- **22.** 구조적 VAR (Structural VAR, SVAR)

## **파트 4: 머신러닝 예측 방법론 (ML for Prediction)**

- **23.** 머신러닝 예측의 세계: 새로운 패러다임
- **24.** Regularization (정규화): 라쏘(Lasso)와 릿지(Ridge)
- **25.** 트리 기반 모델 (1): 랜덤 포레스트 (Random Forest)
- **26.** 트리 기반 모델 (2): 부스팅 (Boosting)
- **27.** 서포트 벡터 머신 (Support Vector Machine, SVM)
- **28.** 딥러닝과 시계열 예측: RNN & LSTM

## **파트 5: 머신러닝 인과추론 (ML for Causal Inference)**

- **29.** 머신러닝 인과추론 핵심 아이디어
- **30.** 이중/편향제거 머신러닝 (Double/Debiased Machine Learning, DML)
- **31.** 머신러닝을 활용한 도구변수법 (IV with Machine Learning)
- **32.** Meta-Learners: 이질적 처치 효과(HTE) 추정 전략
- **33.** 트리 기반 HTE 추정: 일반화 랜덤 포레스트와 인과 포레스트
- **34.** 딥러닝과 인과추론 (Deep Learning & Causal Inference)
- **35.** 인과적 발견 (Causal Discovery)
