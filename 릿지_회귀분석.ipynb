## **논문: Ridge Regression: Biased Estimation for Nonorthogonal Problems(1970)**


흔히 사용하는 최소 제곱 추정법은(OLS)는 예측 변수간에 상관관계가 거의 없을때
좋은 방법이지만, 상관관계가 높다면(다중공선성) 결과가 불안정해질 수 있다. 이를 해결하기 위해 Arthur E. Hoerl과 Robert W. Kennard는 1970년 릿지 회귀분석을 제안하였다. 이 방법은 추정값의 민감도를 쉽게 파악할 수 있고, 더 작은 평균 제곱 오차(MSE)를 가진 추정값을 통해 정확도를 높일 수 있다

일반적인 OLS에서는 변수 간 높은 상관관계가 있다면 매우 작은 고유값(상관관계가 있는 두 변수의 순수한 영향력을 구별 불가능한 상황)으로 분산과 평균오차가 커지게된다. 릿지 회귀는 이를 보완하고자 $X'X$행렬의 대각선에 작은 양수$k$를 더해서 $[X'X+kI]$이라는 새로운 행렬을 만들 고유값을 증가시킨다.

릿지 회귀를 분석할 때는 릿지 트레이스를 사용한다. 릿지 트레이스는 $k$값을 0부터 점차 증가시킬때 회귀계수 값들이 어떻게 변하는지 그래프로 표현한 것이다.  OLS는 잔차 제곱합(RSS)를 최소로 하는 단 하나의 점을 찾지만, 잔차 제곱합을 조금 희생한다면 계수값들의 조합은 무수히 많아진다. 릿지 트레이스는 많은 조합들 중 계수 터(두 계수의 불균형)길이를 가장 짧게 만드는 경로를 나타낸다.

릿지 추정량의 성능은 평균 제곱 오차(MSE)의 관점에서 판단할 수 있다.
$MSE=분산+(편향)^2$이다. $k$가 커짐에 따라 분산은 급격히 감소하고 편향은 서서히 증가한다. 따라서 $k$가 0에 가까운 작은 수 일때, 분산이 감소하는 속도가 편향이 증가하는 속도보다 빠르기 때문에 MSE는 OLS의 오차보다 낮아지는 구간이 반드시 존재하게 된다.

릿지 회귀의 최적 $k$값을 찾는 것은 매우어렵기 때문에 OLS보다 더 나은 값을 찾는 것을 목표로 한다. 릿지 트레이스를 보고 $k$값을 고를 때 가이드 라인을 참고하여야한다.
* 안정성:$k$가 커짐에 따라 계수들이 안정이게 되는 지점을 찾아야한다.
* 계수값의 합리성:회귀 계수들의 절대값이 비정상적으로 크지 않은 수준으로 줄어들어야한다.
* 계수 부호의 합리성:상식과 반대되는 부호가 $k$값이 커짐에 따라 올바르게 바뀌는 지점을 선택해야한다.
* 잔차 제곱의 크기:$k$값을 너무 크게 잡으면 안정적이지만 설명력이없기 때문에 합리적인 지점을 선택해야한다.

## **논문: Ridge Regression: Applications to Nonorthogonal Problems(1970)**

Arthur E. Hoerl과 Robert W. Kennard는 릿지 회귀 분석을 제안하고 실제로 적용하여 분석의 실용성을 입증하였다. 첫 번째는 다른 연구자들이 발표하였던 10개의 변수를 가진 데이터를 활용하였다. 이 데이터의 $X'X$행렬의 가장 작은 고유값이 0.068로 매우 작아 다중공선성 문제가 있음을 알 수있다.

릿지 트레이스를 통하여 기존의 문제점을 확인할 수 있다. $k$가 0일때 계수들이 대평가 되있음을 알 수있는데 특히 5,6번 계수는 $k$값을 조금 증가시킬 때 계수들의 절값이 급격히 줄어드는 것을 확인할 수 있다. 또한 5번 계수는 $k$값을 증가시킬때 부호가 양수로 바뀌는 것을 알 수있다. 5번은 양의 부호인 6번 변수와의 상관계수가 0.84로 높기때문에 두 부호가 정반대일 확률이 낮다.
또한 1번 계수는 $k$가 커짐에 따라 중요해지는데, 이는 과소평가되있는 계수 또한 식별할 수 있음을 알 수 있다. 저자들은 이러한 분석을 바탕으로, 이 모델의 적절한 $k$값은 0.2에서 0.3사이의 값이라고 제안한다.

두번째 예시는 13개의 변수를 가진 광산 지지대 데이터를 가지고 릿지 회귀분석을 실시한다. 13개의 고유값중 마지막 3개는 값이 매우 작아 다중공선성 문제가 심각하다.

스크린샷 2025-08-28 182326.png

이 데이터의 릿지 트레이스를 통하여 계수들의 문제점을 진단할 수 있다. 첫째로 $k$값이 0일 때 계수들의 크기가 지나치 크다. 두번째로 4번과 5번 계수는 $k$값이 증가함에 따라 부호가 바뀌는데 이를 통해 OLS가 이 변수의 영향력을 반대로 해석하였음을 알수있다. 세번째로 2번 계수가 $k$가 0일때는 매우 중요한 변수인 것처럼 보였지만 $k$이 증가할 수록 영향력이 매우 줄어들어 사실은 통계적으로 허상인 것을 알 수있다. 이러한 정보들을 바탕으로 저자는 $k$가 0.2에서 0.4사이에서 안정화된다고 주장한다.

기존에 이 데이터를 분석했던 Jefferss는 주성분 분석(PCA)라는 통계 기법을 사용했다. 저자들은 릿지 회귀와 PCA의 성능을 비교하였는데 릿지 회귀의 설명력이 더 높게 나타나였다.
